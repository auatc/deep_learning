### Conditional selection

```python
data_index.api_element=='Math.incrementExact(long a)'
```

```python
data_index.loc[data_index.api_element=='Math.incrementExact(long a)']
```

```python
# operation symbol:and (&)
data_index.loc[(data_index.api_element=='Math.incrementExact(long a)')&(data_index.cq=='a')]
```

```python
# operation symbol:or(|)
data_index.loc[(data_index.api_element=='Math.incrementExact(long a)')|(data_index.cq=='a')]
```

```python
#DataFrame.isin():lets you select data whose value "is in" a list of values.
data_index.loc[data_index.api_element.isin(['Math.incrementExact(long a)','String.substring(int beginIndex)'])]
```

### Maps

```python
data_index_api_mean=data_index.api_element.mean()
data_index.api_element.map(lambda p:p-data_index_api_mean)
```

```python
def remean_api_element(row):
    row.api_element-=data_index_api_mean
    return row
data_index.apply(remean_api_element,axis='columns')
```

### Groupwise analysis

```python
reviews.head(2)
```

```python
#DataFrame.groupby
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda df:df.title.iloc[0])
```

```python
reviews.groupby(['country','province']).apply(lambda df:df.loc[df.points.idxmax])
```

```python
#agg
reviews.groupby(['country']).price.agg([len,min,max])
```

### Combining

```python
#DataFrame.concat()
canadian_youtube = pd.read_csv("data/CAvideos.csv")
british_youtube = pd.read_csv("data/GBvideos.csv")
pd.concat([canadian_youtube,british_youtube])
```

```python
#DataFrame.join()
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])
left.join(right,lsuffix='_CAN',rsuffix='_UK')
```
